```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./datasets/1/click_stream.csv', names = ['count_clicks', 'page', 'date'])
df.head()
```

```python
df['month'] = pd.DatetimeIndex(df['date']).month
```

```python
statistics = df.groupby(by = ['month', 'page'], as_index=False).agg({'count_clicks':'count'})
statistics
```

```python
home_page = statistics.query('page == "1_home_page"')
search_page = statistics.query('page == "2_search_page"')
payment_page = statistics.query('page == "3_payment_page"')
payment_confirmation_page = statistics.query('page == "4_payment_confirmation_page"')
home_page
```

```python
fig, axis = plt.subplots(2, 2, constrained_layout=True)

axis[0, 0].barh(home_page.month, home_page.count_clicks, height = 0.5)
axis[0, 0].set_title("Home page")

axis[0, 1].barh(search_page.month, search_page.count_clicks, height = 0.5)
axis[0, 1].set_title("Search page")

axis[1, 0].barh(payment_page.month, payment_page.count_clicks, height = 0.5)
axis[1, 0].set_title("Payment page")

axis[1, 1].barh(payment_confirmation_page.month, payment_confirmation_page.count_clicks, height = 0.5)
axis[1, 1].set_title("Payment confirmation page")

plt.show()
```

```python
df = pd.read_csv('./datasets/1/click_stream2.csv', names = ['count_clicks', 'page', 'date', 'device'])
df['month'] = pd.DatetimeIndex(df['date']).month
df.head()
```

```python
statistics = df.groupby(by = ['month', 'page', 'device'], as_index = False).aggregate({'count_clicks': 'count'})
statistics
```

```python
statistics.replace({
    'Desktop' : 'b',
    'Mobile': 'g'
}, inplace= True)
```

```python
home_page = statistics.query('page == "1_home_page"')
search_page = statistics.query('page == "2_search_page"')
payment_page = statistics.query('page == "3_payment_page"')
payment_confirmation_page = statistics.query('page == "4_payment_confirmation_page"')
payment_page
```

```python
fig, axis = plt.subplots(2, 2, constrained_layout=True)

axis[0, 0].scatter(home_page.month, home_page.count_clicks, c = home_page.device, s = 60)
axis[0, 0].set_title("Home page")
axis[0, 0].set_xlabel('month')
axis[0, 0].set_ylabel('count clicks')

axis[0, 1].scatter(search_page.month, search_page.count_clicks, c = search_page.device, s = 60)
axis[0, 1].set_title("Search page")
axis[0, 1].set_xlabel('month')
axis[0, 1].set_ylabel('count clicks')

axis[1, 0].scatter(payment_page.month, payment_page.count_clicks, c = payment_page.device, s = 60)
axis[1, 0].set_title("Payment page")
axis[1, 0].set_xlabel('month')
axis[1, 0].set_ylabel('count clicks')

axis[1, 1].scatter(payment_confirmation_page.month, payment_confirmation_page.count_clicks, c = payment_confirmation_page.device, s = 60)
axis[1, 1].set_title("Payment confirmation page")
axis[1, 1].set_xlabel('month')
axis[1, 1].set_ylabel('count clicks')

fig.legend(["Desktop"], loc = 'center')

plt.show()
```
