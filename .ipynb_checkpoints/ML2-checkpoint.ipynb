```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('./airline_passenger_satisfaction.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.drop('ID', axis=1, inplace=True)
df.head()
```

Решение проблем 0 значений в дата сете:
1) Удалить строки с 0 значением столбца
2) Заменить 0 на среднее значение столбца
3) Заменить 0 на что-либо еще

```python
df['Arrival Delay'].mean()
```

```python
df.fillna({'Arrival Delay':df['Arrival Delay'].mean()}, inplace=True)
```

Ручное кодирование (следует избегать use 1/0 тк некоторые модели могут воспринять это как bool type)

```python
df.replace({
    'Gender':{
        'Male':1,
        'Female':2
    },
    'Customer Type':{
        'First-time':1,
        'Returning':2
    },
    'Type of Travel':{
        'Business':1,
        'Personal':2
    },
    'Class':{
        'Business':1,
        'Economy':2,
        'Economy Plus':3
    }
})
```

```python
label_encoder = LabelEncoder()
columns = df.select_dtypes(include='object').drop(columns='Satisfaction').columns

for column in columns:
    df[column] = label_encoder.fit_transform(df[column])

df.head()
```

```python
df.dtypes
```

```python
df.loc[df['Age']> 50, ['Gender', 'Class', 'Flight Distance']].head()
```

```python
X = df.drop(columns='Satisfaction')
y = df['Satisfaction']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
p = model.predict(X_test)
score = accuracy_score(y_test, p)
score
```

#### Random Forest

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model1 = RandomForestClassifier()
model1.fit(X_train, y_train)
```

```python
p = model1.predict(X_test)
score = accuracy_score(y_test, p)
score
```

#### KNeighbors

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

```python
p = model.predict(X_test)
score = accuracy_score(y_test, p)
score
```

#### Logistic Regression

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter = 100)
model.fit(X_train, y_train)
```

```python
p = model.predict(X_test)
score = accuracy_score(y_test, p)
score
```
