```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import seaborn as sns
```

# Loading Data

```python
df = pd.read_csv('./vehicles.csv')
df.head(4)
```

```python
df.dtypes
```

# Cleaning

Нельзя, чтобы в данных были пустые занчения

```python
df.isnull().sum()
```

```python
df2 = pd.DataFrame(df)
df2['Income'].fillna(0.0, inplace=True)
df2
```

```python
df2.isnull().sum()
```

# Encoding

Нужно закодировать male и female так-как это входные параметры для обучения и str не пойдет

```python
df3 = pd.DataFrame(df2)
df4 = pd.DataFrame(df2)
df3.head(3)
```

```python
df3['Gender'].unique()
```

### Так-как значений не много можно сделать ручное кодирование

```python
df3.replace({
    'Gender':{
        'male':0,
        'female':1
    }
}, inplace=True)
df3
```

```python
df3.dtypes
```

### Другой вариант, когда заменить нужно много значений

```python
label_encoder = LabelEncoder()
df4['Gender'] = label_encoder.fit_transform(df2['Gender'])
df4
```

# Model

```python
# Входные данные
X = df4.drop(columns='Favorite Transport')
```

```python
# Целевые показатели
y = df4['Favorite Transport']
y.head(3)
```

```python
model = DecisionTreeClassifier()
model.fit(X,y) # Запуск обучения
```

# Prediction

```python
test_df = pd.DataFrame({
    'Age':[68],
    'Gender': [1],
    'Income': [5.5]
})
test_df
```

```python
model.predict(test_df) #На основе данных модель предсказывает целевое показатель
```

# Evaluation

Тест модели происходит таким образом, что часть df мы отдаем на обучение, а часть нет. Ту часть которую мы не отдали послужит проверкой на точность ответов, так-как мы точно знаем, какой должен быть ответ. Далее идет сравнение результатов, которые дала модель с df которые не пошли на обучение.

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)#20% для тестирования, 80% для обучения

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
X_test
```

```python
predictions = model.predict(X_test)
predictions
```

Чем больше тестов, тем высше шанс ошибки модели => меньше % точности

Модель считается хорошей, если ее точность выше 80%

Данные должны быть реальными, а не случайными. Если данные случайные и нет строгой зависимости между вводными данными и целевыми показателями, то точность модели будет очень низкой. Мы как-раз и создаем модель, чтобы находить зависимость(закон соотношения) между вводными данными и целевыми показателями, а если эта зависимость случайна, то модель не имеет смысла.

```python
accuracy_score(y_test, predictions)
```

# DOT File

```python
tree.export_graphviz(model, out_file='tree_model.dot', feature_names=['Age', 'Gender', 'Income'], filled=True, class_names=sorted(y.unique()))
```
